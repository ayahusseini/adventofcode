## The task

- claim ticket said '2017 JP' 
- Item brought in in the year 2017
- Initials of the contracter are 'JP' 

```python
query = """
SELECT *
FROM customers AS c
INNER JOIN orders o ON c.customerid = o.customerid
LEFT OUTER JOIN orders_items oi on oi.orderid = o.orderid
LEFT JOIN products p on p.sku = oi.sku
WHERE strftime("%Y",o.ordered) == '2017'
AND c.name LIKE 'J% P%'
AND p.desc LIKE '%Cleaner%'
;
"""
```

```python
pd.read_sql(query,conn)
```
